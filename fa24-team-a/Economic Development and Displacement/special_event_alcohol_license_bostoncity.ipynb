# Special event alcohol licenses 
This dataset provides  insights into alcohol licensing in Boston City, with a filtered subset for District 7. The analysis focuses on various dimensions of the licenses to uncover trends, patterns, and disparities between District 7 and Boston City. 

```python
alcohol_licenses_df = pd.read_csv('cleaned_data.csv')

census_tracts_list = [
    '080601', '010403', '070700', '010300', '981100',
    '120201', '980300', '010404', '010405', '080100',
    '120301', '071101', '091300', '010500', '110105',
    '010206', '081301', '081302', '070502', '082100',
    '010600', '070600', '080300', '080401', '080500',
    '080801', '080900', '081400', '081500', '081700',
    '081800', '081900', '082000', '090100', '090200',
    '090300', '090400', '090600', '090700', '091400',
    '092400', '070902', '070802', '070801', '070901'
]

df_D7 = alcohol_licenses_df[alcohol_licenses_df['census tract'].astype(str).isin(census_tracts_list)]
df_D7.head()
```

# License type distribution


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the full dataset (Boston city data)
alcohol_licenses_df = pd.read_csv('cleaned_data.csv')

# Load District 7 data (already filtered)
df_D7 = alcohol_licenses_df[alcohol_licenses_df['census tract'].astype(str).isin(census_tracts_list)]

# Category to compare: License Types
# Count the number of occurrences for each license type in both datasets
license_type_count_city = alcohol_licenses_df['license_type'].value_counts()
license_type_count_D7 = df_D7['license_type'].value_counts()

# Create side-by-side plots using subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot the distribution for the whole city (Boston)
sns.barplot(x=license_type_count_city.index, y=license_type_count_city.values, palette='viridis', ax=axes[0])
axes[0].set_title('License Type Distribution - Boston City', fontsize=14)
axes[0].set_xlabel('License Type', fontsize=12)
axes[0].set_ylabel('Count', fontsize=12)
axes[0].tick_params(axis='x', rotation=45)

# Plot the distribution for District 7
sns.barplot(x=license_type_count_D7.index, y=license_type_count_D7.values, palette='viridis', ax=axes[1])
axes[1].set_title('License Type Distribution - District 7', fontsize=14)
axes[1].set_xlabel('License Type', fontsize=12)
axes[1].set_ylabel('Count', fontsize=12)
axes[1].tick_params(axis='x', rotation=45)

# Adjust layout for better visualization
plt.tight_layout()
plt.show()
```

# Licenses issued overtime


```python
# Convert 'issued' column to datetime
alcohol_licenses_df['issued'] = pd.to_datetime(alcohol_licenses_df['issued'], errors='coerce')
df_D7['issued'] = pd.to_datetime(df_D7['issued'], errors='coerce')

# Extract year and count licenses per year
alcohol_licenses_df['year'] = alcohol_licenses_df['issued'].dt.year
df_D7['year'] = df_D7['issued'].dt.year

licenses_by_year_city = alcohol_licenses_df.groupby('year').size()
licenses_by_year_d7 = df_D7.groupby('year').size()

# Plot side-by-side
fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

licenses_by_year_d7.plot(kind='bar', color='skyblue', ax=axes[0], edgecolor='black')
axes[0].set_title("District 7: Licenses Issued Over Time")
axes[0].set_xlabel("Year")
axes[0].set_ylabel("Number of Licenses")

licenses_by_year_city.plot(kind='bar', color='orange', ax=axes[1], edgecolor='black')
axes[1].set_title("Boston City: Licenses Issued Over Time")
axes[1].set_xlabel("Year")

plt.tight_layout()
plt.show()
```

# Top 10 business in alcohol business

```python
top_businesses_city = alcohol_licenses_df['business_name'].value_counts().head(10)
top_businesses_d7 = df_D7['business_name'].value_counts().head(10)

fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

sns.barplot(x=top_businesses_d7.values, y=top_businesses_d7.index, palette='coolwarm', ax=axes[0])
axes[0].set_title("Top 10 Businesses: District 7")
axes[0].set_xlabel("Count")
axes[0].set_ylabel("Business Name")

sns.barplot(x=top_businesses_city.values, y=top_businesses_city.index, palette='coolwarm', ax=axes[1])
axes[1].set_title("Top 10 Businesses: Boston City")
axes[1].set_xlabel("Count")
axes[1].set_ylabel("")

plt.tight_layout()
plt.show()
```

# Top 10 areas with licenses


```python
licenses_by_city = alcohol_licenses_df['city'].value_counts().head(10)
licenses_by_city_d7 = df_D7['city'].value_counts().head(10)

fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

sns.barplot(x=licenses_by_city_d7.values, y=licenses_by_city_d7.index, palette='coolwarm', ax=axes[0])
axes[0].set_title("Top 10 Cities: District 7")
axes[0].set_xlabel("Count")
axes[0].set_ylabel("City")

sns.barplot(x=licenses_by_city.values, y=licenses_by_city.index, palette='coolwarm', ax=axes[1])
axes[1].set_title("Top 10 Cities: Boston City")
axes[1].set_xlabel("Count")
axes[1].set_ylabel("")

plt.tight_layout()
plt.show()
```

# Top 10 managers with alcohol licenses

```python
licenses_by_manager_city = alcohol_licenses_df['manager'].value_counts().head(10)
licenses_by_manager_d7 = df_D7['manager'].value_counts().head(10)

fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

sns.barplot(x=licenses_by_manager_d7.values, y=licenses_by_manager_d7.index, palette='Blues_r', ax=axes[0])
axes[0].set_title("Top 10 Managers: District 7")
axes[0].set_xlabel("Number of Licenses")
axes[0].set_ylabel("Manager")

sns.barplot(x=licenses_by_manager_city.values, y=licenses_by_manager_city.index, palette='Blues_r', ax=axes[1])
axes[1].set_title("Top 10 Managers: Boston City")
axes[1].set_xlabel("Number of Licenses")
axes[1].set_ylabel("")

plt.tight_layout()
plt.show()
```

```python
numeric_columns_city = alcohol_licenses_df.select_dtypes(include=['float64', 'int64'])
numeric_columns_d7 = df_D7.select_dtypes(include=['float64', 'int64'])

correlation_matrix_city = numeric_columns_city.corr()
correlation_matrix_d7 = numeric_columns_d7.corr()

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.heatmap(correlation_matrix_d7, annot=True, cmap='coolwarm', fmt='.2f', ax=axes[0])
axes[0].set_title("Correlation Heatmap: District 7")

sns.heatmap(correlation_matrix_city, annot=True, cmap='coolwarm', fmt='.2f', ax=axes[1])
axes[1].set_title("Correlation Heatmap: Boston City")

plt.tight_layout()
plt.show()
```

# Licenses issued every month

```python
# Extract month and count licenses
alcohol_licenses_df['month'] = alcohol_licenses_df['issued'].dt.month
df_D7['month'] = df_D7['issued'].dt.month

licenses_by_month_city = alcohol_licenses_df['month'].value_counts().sort_index()
licenses_by_month_d7 = df_D7['month'].value_counts().sort_index()

fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

licenses_by_month_d7.plot(kind='bar', color='skyblue', ax=axes[0], edgecolor='black')
axes[0].set_title("District 7: Licenses Issued by Month")
axes[0].set_xlabel("Month")
axes[0].set_ylabel("Count")

licenses_by_month_city.plot(kind='bar', color='orange', ax=axes[1], edgecolor='black')
axes[1].set_title("Boston City: Licenses Issued by Month")
axes[1].set_xlabel("Month")

plt.tight_layout()
plt.show()
```

# License type proportion comparison

```python
# Calculate proportions
license_type_prop_city = alcohol_licenses_df['license_type'].value_counts(normalize=True)
license_type_prop_d7 = df_D7['license_type'].value_counts(normalize=True)

fig, axes = plt.subplots(1, 2, figsize=(14, 7))

license_type_prop_d7.plot(kind='pie', autopct='%1.1f%%', ax=axes[0], cmap='coolwarm', legend=True)
axes[0].set_title("District 7: License Type Proportion")
axes[0].set_ylabel("")  # Hide y-label for aesthetics

license_type_prop_city.plot(kind='pie', autopct='%1.1f%%', ax=axes[1], cmap='coolwarm', legend=True)
axes[1].set_title("Boston City: License Type Proportion")
axes[1].set_ylabel("")

plt.tight_layout()
plt.show()
```

# Manager vs the license type they hold

```python
# Group by manager and license type
manager_license_city = alcohol_licenses_df.groupby(['manager', 'license_type']).size().unstack(fill_value=0).head(10)
manager_license_d7 = df_D7.groupby(['manager', 'license_type']).size().unstack(fill_value=0).head(10)

fig, axes = plt.subplots(1, 2, figsize=(14, 8))

sns.heatmap(manager_license_d7, cmap='Blues', annot=True, fmt='d', ax=axes[0])
axes[0].set_title("District 7: Manager vs. License Type")

sns.heatmap(manager_license_city, cmap='Blues', annot=True, fmt='d', ax=axes[1])
axes[1].set_title("Boston City: Manager vs. License Type")

plt.tight_layout()
plt.show()
```
