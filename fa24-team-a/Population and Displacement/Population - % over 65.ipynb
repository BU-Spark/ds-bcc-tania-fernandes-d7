*The primary objective of this analysis is to examine trends in the percentage of individuals aged 65 and older across different regions.*

The dataset analyzed here consists of census data collected from 2014 to 2023, focusing on the population distribution by age. By examining these trends over time, we aim to gain a clearer understanding of how aging populations may be affected by demographic shifts and displacement risks.

#### Plotting the trends for total population over 65 along with a gender-wise distribution 

```python
percentage_data = over65.iloc[[0, 2, 3]].copy()

# Remove the '%' sign and convert to numeric values
for col in percentage_data.columns[1:]:
    percentage_data[col] = percentage_data[col].str.rstrip('%').astype(float)

# Set the labels as the index
percentage_data.set_index('Labels', inplace=True)

# Transpose the dataframe for plotting
percentage_data = percentage_data.T
```

```python
# Reordering the columns to display the years in ascending order
percentage_data = percentage_data.sort_index(ascending=True)

# Plot for '65 years and over' (Total Population)
plt.figure(figsize=(8, 5))
sns.lineplot(data=percentage_data['65 years and over'], marker='o', color='green')
plt.title('Trend of Total Population 65 Years and Over (Percentage)')
plt.xlabel('Year')
plt.ylabel('Percentage (Total)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

last_two_columns = percentage_data.iloc[:, -2:]

plt.figure(figsize=(8, 5))
for column in last_two_columns.columns:
    plt.plot(last_two_columns.index, last_two_columns[column], marker='o', label=column)

# Add titles and labels
plt.title('Trend for Last Two Columns')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend(title='Columns')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```
