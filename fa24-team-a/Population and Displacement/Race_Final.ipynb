## Summary

This notebook delves into population dynamics across various racial and ethnic groups, providing a comprehensive analysis of demographic trends over time. By leveraging publicly available census data, the study investigates key aspects of diversity and representation within the population. The following comparisons have been conducted:

1. **One Race vs. Two or More Races**:  
   Analyzing the prevalence and growth of individuals identifying with a single race compared to those identifying with multiple races. This comparison highlights the increasing diversity in racial identities.

2. **Asian Population Distribution**:  
   A focused look at the Asian population, analyzing trends within this group and understanding their distribution and growth patterns over time.

3. **Hispanic Population Distribution**:  
   A detailed exploration of Hispanic demographic trends, shedding light on their representation and how this group’s population has changed across different periods.

4. **Kinds of One Race Population**:  
   A breakdown of the population identifying with one race, segmented into specific racial categories (e.g., Black or African American, Asian, White). This section offers a nuanced understanding of the individual groups within the broader "one race" category and their population trends.

5. **White vs. Non-White**:  
   Examining the demographic shifts between White and Non-White populations, providing insights into how the racial composition of the population is evolving.   

The notebook not only captures these overarching trends but also conducts a focused comparison of **District 7** with the broader citywide averages. By analyzing population trends within this specific district, the study reveals critical regional disparities, offering a lens into how demographic factors may vary across localities.

This analysis provides a dual perspective: a macro-level view of citywide trends and a micro-level insight into District 7, facilitating a nuanced understanding of population dynamics. Such insights are essential for policymakers, urban planners, and community leaders aiming to address regional needs effectively.

The dataset used for this analysis is sourced from the United States Census Bureau and is publicly accessible at: [Census Data](https://data.census.gov/table?q=demographics&g=050XX00US36103_160XX00US4159000).

## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Pre-processing data for years - 2022 to 2013

Loading the pre-processed dataset containing data for all census tracts associated with District 7, covering the years 2013 through 2022.

```python
df = pd.read_csv('census_analysis_results.csv')
df
```

#### Using this dataset, we will generate graphs to visualize population trends across various racial and ethnic groups. Additionally, the analysis will include a detailed comparison of the population trends between White and Non-White groups.

## One Race vs Two or More Races

```python
one_vs_two_or_more = df[df['label'].isin(['One race', 'Two or more races', 'Two or More Races'])]
one_vs_two_or_more
```

### Graph for Trend of Population for One Race vs Two or More Races for District 7

```python
# Filter data for "One race" and "Two or more races"
one_race = one_vs_two_or_more[one_vs_two_or_more['label'] == 'One race']
two_or_more_races = one_vs_two_or_more[one_vs_two_or_more['label'].isin(['Two or more races', 'Two or More Races'])]

# Treat years as categorical to ensure even spacing
all_years = list(range(one_vs_two_or_more['year'].min(), one_vs_two_or_more['year'].max() + 1))

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(one_race['year'], one_race['percentage'], label='One race', marker='o')
plt.plot(two_or_more_races['year'], two_or_more_races['percentage'], label='Two or more races', marker='o')

# Use categorical ticks for even spacing
plt.xticks(all_years, rotation=45)  # Set all years on the x-axis

# Customize the plot
plt.title('Trend of Population for One Race vs Two or More Races', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Percentage (%)', fontsize=12)
plt.legend(title='Race Category', fontsize=10)
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

# White vs Non-White

## White

```python
white_total = df[df['label'].isin(['White 2'])]
white_total
```

## Non-White

```python
non_white = df[df['label'].isin(['Black or African American 2', 'American Indian and Alaska Native 2', 'Asian 2', 
                                 'Native Hawaiian and Other Pacific Islander 2','Some other race 2', 'Some Other Race 2',
                                 'Hispanic or Latino (of any race)'])]
non_white['label'] = non_white['label'].copy().replace('Some Other Race 2', 'Some other race 2')
non_white
```

### Adding all the categories to create Non-White Dataset

```python
# Group by 'year' and sum numeric columns
non_white_total = non_white.groupby('year', as_index=False).sum()
non_white_total['label'] = 'Non White'

# Display the result
non_white_total
```

### Graph for Trend of Population for White vs Non-White for District 7

```python
# Concatenate the datasets
combined_data = pd.concat([white_total, non_white_total])

# Group by year and label, summing percentages
summary = combined_data.groupby(['year', 'label'], as_index=False)['percentage'].sum()

# Plot the data
plt.figure(figsize=(10, 8))
sns.lineplot(data=summary, x="year", y="percentage", hue="label", marker="o")

# Customize the plot
all_years = list(range(combined_data['year'].min(), combined_data['year'].max() + 1))
plt.xticks(all_years, rotation=45)
plt.title("Percentage Trends: White vs Non-White (Yearly)", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Percentage", fontsize=14)
plt.legend(title="Label")
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```
