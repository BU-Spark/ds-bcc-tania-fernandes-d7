### Replacing the column headers with the first row (column definitions)

```python
# List of dataframes
dataframes = [df_2011, df_2012, df_2013, df_2014, df_2015, df_2016, df_2017, df_2018, df_2019, df_2020, df_2021, df_2022]

# Function to replace column headers with the first row
def replace_headers(df):
    df.columns = df.iloc[0]  # Set the first row as the header
    df = df.drop(index=0)  # Drop the first row
    df.reset_index(drop=True, inplace=True)  # Reset the index
    return df

df_2011 = replace_headers(df_2011)
df_2012 = replace_headers(df_2012)
df_2013 = replace_headers(df_2013)
df_2014 = replace_headers(df_2014)
df_2015 = replace_headers(df_2015)
df_2016 = replace_headers(df_2016)
df_2017 = replace_headers(df_2017)
df_2018 = replace_headers(df_2018)
df_2019 = replace_headers(df_2019)
df_2020 = replace_headers(df_2020)
df_2021 = replace_headers(df_2021)
df_2022 = replace_headers(df_2022)
```

### Removing the margin of error columns for easier understanding

```python
def remove_margin_of_error_columns(df):
    # Identify columns to drop
    columns_to_drop = df.filter(like='Margin of Error').columns
    # Drop the identified columns
    df = df.drop(columns=columns_to_drop)
    return df

df_2011 = remove_margin_of_error_columns(df_2011)
df_2012 = remove_margin_of_error_columns(df_2012)
df_2013 = remove_margin_of_error_columns(df_2013)
df_2014 = remove_margin_of_error_columns(df_2014)
df_2015 = remove_margin_of_error_columns(df_2015)
df_2016 = remove_margin_of_error_columns(df_2016)
df_2017 = remove_margin_of_error_columns(df_2017)
df_2018 = remove_margin_of_error_columns(df_2018)
df_2019 = remove_margin_of_error_columns(df_2019)
df_2020 = remove_margin_of_error_columns(df_2020)
df_2021 = remove_margin_of_error_columns(df_2021)
df_2022 = remove_margin_of_error_columns(df_2022)
```
