#### Removing columns that we are not concerned with, aggregating the dataframes such that there is 1 single aggregated row for each year with the total / median information

```python
import pandas as pd
import numpy as np

def convert_columns_to_numeric(df):
    return df.apply(pd.to_numeric, errors='coerce')

def remove_specific_columns(df):
    columns_to_drop = df.filter(regex='Geograph|PERCENT|Margin of Error|Percent|Moved In|Moved in|SELECTED MONTHLY OWNER COSTS|YEAR STRUCTURE BUILT|VEHICLES AVAILABLE|HOUSE HEATING FUEL|YEAR HOUSEHOLDER MOVED INTO UNIT|complete plumbing facilities|MORTGAGE STATUS|OCCUPANTS PER ROOM|kitchen').columns
    columns_to_keep = [col for col in df.columns if col not in columns_to_drop]
    if len(columns_to_keep) == 0:
        print("Warning: All columns would be dropped. Keeping all columns instead.")
        return df
    return df[columns_to_keep]

def aggregate_rows(df):
    # Identify median columns
    median_columns = df.filter(regex='Median').columns
    
    # Aggregate non-median columns by sum
    sum_df = df.drop(columns=median_columns).sum().to_frame().T
    
    # Aggregate median columns by mean
    median_df = df[median_columns].mean().to_frame().T
    
    # Combine the results
    return pd.concat([sum_df, median_df], axis=1)

def process_dataframe(df):
    df = convert_columns_to_numeric(df)
    df = remove_specific_columns(df)
    df = aggregate_rows(df)
    return df

# Apply the process_dataframe function to each dataframe manually
df_2011_d7 = process_dataframe(df_2011_d7)
df_2012_d7 = process_dataframe(df_2012_d7)
df_2013_d7 = process_dataframe(df_2013_d7)
df_2014_d7 = process_dataframe(df_2014_d7)
df_2015_d7 = process_dataframe(df_2015_d7)
df_2016_d7 = process_dataframe(df_2016_d7)
df_2017_d7 = process_dataframe(df_2017_d7)
df_2018_d7 = process_dataframe(df_2018_d7)
df_2019_d7 = process_dataframe(df_2019_d7)
df_2020_d7 = process_dataframe(df_2020_d7)
df_2021_d7 = process_dataframe(df_2021_d7)
df_2022_d7 = process_dataframe(df_2022_d7)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

for year in range(2011, 2023):
    df = globals()[f'df_{year}']
    df['Median Rent($)'] = pd.to_numeric(df['Median Rent($)'], errors='coerce')

# Create a list of years
years = range(2011, 2023)

# Create lists to store the median rent values
d7_median_rents = []
boston_median_rents = []

# Extract median rent values from each dataframe
for year in years:
    d7_df = globals()[f'df_{year}_d7']
    boston_df = globals()[f'df_{year}']
    
    d7_median_rents.append(d7_df['Median Rent($)'].values[0])
    boston_median_rents.append(boston_df['Median Rent($)'].values[0])

# Create the plot
plt.figure(figsize=(12, 6))
plt.plot(years, d7_median_rents, marker='o', label='District 7')
plt.plot(years, boston_median_rents, marker='s', label='Boston')

plt.title('Median Rent Comparison: District 7 vs Boston (2011-2022)')
plt.xlabel('Year')
plt.ylabel('Median Rent ($)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# Add value labels
for i, (d7_rent, boston_rent) in enumerate(zip(d7_median_rents, boston_median_rents)):
    plt.text(years[i], d7_rent, f'${d7_rent:,.0f}', ha='left', va='bottom')
    plt.text(years[i], boston_rent, f'${boston_rent:,.0f}', ha='right', va='top')

plt.tight_layout()
plt.show()
```
