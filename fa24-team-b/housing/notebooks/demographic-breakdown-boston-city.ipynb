```python
import pandas as pd

# Function to remove the last column of the dataframe
def remove_last_column(df):
    df = df.iloc[:, :-1]
    return df

# Function to replace column names with the first row of the dataframe
def replace_column_names_with_first_row(df):
    df.columns = df.iloc[0]
    df = df[1:]
    return df

# Function to convert all columns with numbers in them to numeric
def convert_columns_to_numeric(df):
    df = df.apply(pd.to_numeric, errors='coerce')
    return df

# Function to remove columns containing specific words in the column name
def remove_specific_columns(df):
    columns_to_drop = df.filter(regex='Geograph|PERCENT|Margin of Error|Percent|Moved In|Moved in').columns
    df = df.drop(columns=columns_to_drop)
    return df

# Function to aggregate all rows of each column to sum them
def aggregate_rows(df):
    aggregated_df = df.sum(numeric_only=True).to_frame().T
    return aggregated_df

# Function to process the dataframe according to the specified steps
def process_dataframe(df):
    df = remove_last_column(df)
    df = replace_column_names_with_first_row(df)
    df = convert_columns_to_numeric(df)
    df = remove_specific_columns(df)
    df = aggregate_rows(df)
    return df


# Apply the processing function to each dataframe
df_2011_processed = process_dataframe(df_2011)
df_2012_processed = process_dataframe(df_2012)
df_2013_processed = process_dataframe(df_2013)
df_2014_processed = process_dataframe(df_2014)
df_2015_processed = process_dataframe(df_2015)
df_2016_processed = process_dataframe(df_2016)
df_2017_processed = process_dataframe(df_2017)
df_2018_processed = process_dataframe(df_2018)
df_2019_processed = process_dataframe(df_2019)
df_2020_processed = process_dataframe(df_2020)
df_2021_processed = process_dataframe(df_2021)
df_2022_processed = process_dataframe(df_2022)

print("Processing complete for all dataframes.")
```

### Separating By Housing Type, such that instead of separate columns for "Occupied - Under 35", "Owner Occupied - Under 35" and "Renter Occupied - Under 35" for the year 2011, I will have 3 rows for the year 2011, one for each type of housing (occupied, owner occupied, renter occupied)

```python
import pandas as pd

df = pd.read_excel('processed_data.xlsx')
```

```python
df_melted = df.melt(id_vars=['Year'], var_name='Category', value_name='Value')
```

```python
def determine_housing_type(category):
    if category.startswith('Owner Occupied'):
        return 'Owner Occupied'
    elif category.startswith('Renter Occupied'):
        return 'Renter Occupied'
    else:
        return 'Occupied'

# Apply the function to create the new 'Housing Type' column
df_melted['Housing Type'] = df_melted['Category'].apply(determine_housing_type)
```

```python
categories = set()
for col in df.columns:
    if col != 'Year' and ':' in col:
        categories.add(col.split(': ', 1)[1])

# Function to reshape the dataframe
def reshape_dataframe(df):
    reshaped_data = []
    for _, row in df.iterrows():
        year = row['Year']
        # Define totals for each housing type
        totals = {
            'Occupied': row['Occupied Housing Units'],
            'Owner Occupied': row['Owner Occupied Housing Units'],
            'Renter Occupied': row['Renter Occupied Housing Units']
        }
        for housing_type in ['Occupied', 'Owner Occupied', 'Renter Occupied']:
            new_row = {'Year': year, 'Housing Type': housing_type, 'Total': totals[housing_type]}
            for category in categories:
                col_name = f"{housing_type}: {category}"
                if col_name in df.columns:
                    new_row[category] = row[col_name]
                else:
                    new_row[category] = None
            reshaped_data.append(new_row)
    return pd.DataFrame(reshaped_data)

# Reshape the dataframe
reshaped_df = reshape_dataframe(df)

# Reorder columns
column_order = ['Year', 'Housing Type', 'Total'] + sorted(list(categories))
reshaped_df = reshaped_df[column_order]

# Display the reshaped dataframe
reshaped_df
```
