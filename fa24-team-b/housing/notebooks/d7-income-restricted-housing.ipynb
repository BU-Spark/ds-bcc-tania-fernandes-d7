```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("../data/income-restricted-housing-inventory-2022.csv")
data
```

#### Clean neighborhood column
- Fenway are listed twice, one as their own entity (`Fenway`) and one as a part of a bigger administrative body (`Fenway/Kenmore`).
- `South End` and `South end` both exist.

Turn the data into all lowercase and retain all fields that contain the neighborhood's name.

```python
unique_neighborhood = data['Neighborhood'].unique()
np.sort(unique_neighborhood)
```

```python
data['Neighborhood'] = data['Neighborhood'].str.lower()
data
```

```python
d7_neighborhood_list = ['roxbury', 'dorchester', 'fenway', 'south end']
d7_neighborhood_data = data[data['Neighborhood'].isin(d7_neighborhood_list)]
d7_neighborhood_data
```

#### Examine the neighborhoods with obscurity about whether it falls under District 7 juridistiction.
- South End
- Fenway and Fenway/Kenmore

##### South End
- Manually find out which district the project belongs to from the Find My Councilor tool.
- And additional column called `D7` is appended to the CSV file to indicate whether a particular project is located in District 7. The value is 1 if the project is in D7, 0 if it doesn't, and ??? if the address cannot be determined.

```python
south_end_income_restricted_housings = pd.read_csv("../data/south-end-restricted-income-housing.csv")
D7_south_end_housings = south_end_income_restricted_housings[south_end_income_restricted_housings['D7'] == '1']
D7_south_end_housings = D7_south_end_housings.drop(columns=['D7'])
D7_south_end_housings
```

#### Fenway
- There is only 1 ambiguous project whose neighborhood is listed as `Fenway/Kenmore`. After using the Find My Councilor tool, the project is found to not be located in District 7, so we will remove the particular row.

```python
data = data[data['Neighborhood'] != "fenway/kenmore"]
data
```

#### Finalize that dataset for income restricted housings in District 7

```python
D7_income_restricted_housing = data[data['Neighborhood'].isin(['roxbury', 'fenway', 'dorchester'])]
D7_income_restricted_housing = pd.concat([D7_income_restricted_housing, D7_south_end_housings])
D7_income_restricted_housing
```

#### Export to CSV

```python
D7_income_restricted_housing.to_csv("../data/d7-income-restricted-housing.csv", index=False)
```
