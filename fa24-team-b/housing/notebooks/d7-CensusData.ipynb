## United States Census Bureau Dataset

This dataset contains housing and demographic information for geographic areas for the years 2011 to 2022. It includes various attributes like information on housing occupancy (total units, occupied and vacant), the number of units in each structure, the number of rooms and bedrooms in each structure, the housing tenure information, value and gross rent. More information can be found here. For this project, we are focusing on zipcodes 02119 and 02121 (District 7).

This dataset will also be utilized to compare the income restricted housing statistics for District 7, against the housing details for District 7 as a whole.

### Visualizing the Housing Occupancy Over Time

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming the datasets are already loaded and processed
# List of dataframes for each year
dfs = [df_2011, df_2012, df_2013, df_2014, df_2015, df_2016, df_2017, df_2018, df_2019, df_2020, df_2021, df_2022]

# List of years corresponding to the dataframes
years = list(range(2011, 2023))

# Initialize dictionaries to store the data for each region
data_02119 = {'Year': [], 'Total Housing Units': [], 'Occupied Housing Units': [], 'Vacant Housing Units': []}
data_02121 = {'Year': [], 'Total Housing Units': [], 'Occupied Housing Units': [], 'Vacant Housing Units': []}

# Iterate over each dataframe and extract the relevant data
for year, df in zip(years, dfs):
    if 'Geographic Area Name' in df.columns and 'Total Housing Units' in df.columns and 'Occupied Housing Units' in df.columns and 'Vacant Housing Units' in df.columns:
        for index, row in df.iterrows():
            if row['Geographic Area Name'] == 'ZCTA5 02119':
                data_02119['Year'].append(year)
                data_02119['Total Housing Units'].append(pd.to_numeric(row['Total Housing Units'], errors='coerce'))
                data_02119['Occupied Housing Units'].append(pd.to_numeric(row['Occupied Housing Units'], errors='coerce'))
                data_02119['Vacant Housing Units'].append(pd.to_numeric(row['Vacant Housing Units'], errors='coerce'))
            elif row['Geographic Area Name'] == 'ZCTA5 02121':
                data_02121['Year'].append(year)
                data_02121['Total Housing Units'].append(pd.to_numeric(row['Total Housing Units'], errors='coerce'))
                data_02121['Occupied Housing Units'].append(pd.to_numeric(row['Occupied Housing Units'], errors='coerce'))
                data_02121['Vacant Housing Units'].append(pd.to_numeric(row['Vacant Housing Units'], errors='coerce'))
    else:
        print(f"Missing expected columns in dataframe for year {year}")

# Convert dictionaries to dataframes
df_02119 = pd.DataFrame(data_02119)
df_02121 = pd.DataFrame(data_02121)

# Sort the dataframes by Year to ensure correct order
df_02119 = df_02119.sort_values(by='Year')
df_02121 = df_02121.sort_values(by='Year')

# Plot the data
plt.figure(figsize=(14, 8))

# Plot for ZCTA5 02119
plt.subplot(2, 1, 1)
plt.plot(df_02119['Year'], df_02119['Total Housing Units'], label='Total Housing Units', color='blue', marker='o')
plt.plot(df_02119['Year'], df_02119['Occupied Housing Units'], label='Occupied Housing Units', color='orange', marker='o')
plt.plot(df_02119['Year'], df_02119['Vacant Housing Units'], label='Vacant Housing Units', color='green', marker='o')
plt.title('Housing Units Over Time for ZCTA5 02119')
plt.xlabel('Year')
plt.ylabel('Number of Units')
plt.legend()
plt.ylim(min(df_02119[['Total Housing Units', 'Occupied Housing Units', 'Vacant Housing Units']].min()) - 100, 
         max(df_02119[['Total Housing Units', 'Occupied Housing Units', 'Vacant Housing Units']].max()) + 100)

# Plot for ZCTA5 02121
plt.subplot(2, 1, 2)
plt.plot(df_02121['Year'], df_02121['Total Housing Units'], label='Total Housing Units', color='blue', marker='o')
plt.plot(df_02121['Year'], df_02121['Occupied Housing Units'], label='Occupied Housing Units', color='orange', marker='o')
plt.plot(df_02121['Year'], df_02121['Vacant Housing Units'], label='Vacant Housing Units', color='green', marker='o')
plt.title('Housing Units Over Time for ZCTA5 02121')
plt.xlabel('Year')
plt.ylabel('Number of Units')
plt.legend()
plt.ylim(min(df_02121[['Total Housing Units', 'Occupied Housing Units', 'Vacant Housing Units']].min()) - 100, 
         max(df_02121[['Total Housing Units', 'Occupied Housing Units', 'Vacant Housing Units']].max()) + 100)

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```

### Deleting some more unnecessary columns

```python
import pandas as pd

def delete_columns_containing(df, keyword):
    # Ensure the columns are of string type
    df.columns = df.columns.astype(str)
    # Delete columns that contain the specified keyword
    df = df.loc[:, ~df.columns.str.contains(keyword)]
    return df

# Apply the delete_columns_containing function to each dataset
df_2011 = delete_columns_containing(df_2011, "GROSS RENT AS A PERCENTAGE OF HOUSEHOLD INCOME")
df_2012 = delete_columns_containing(df_2012, "GROSS RENT AS A PERCENTAGE OF HOUSEHOLD INCOME")
df_2013 = delete_columns_containing(df_2013, "GROSS RENT AS A PERCENTAGE OF HOUSEHOLD INCOME")
df_2014 = delete_columns_containing(df_2014, "GROSS RENT AS A PERCENTAGE OF HOUSEHOLD INCOME")
df_2015 = delete_columns_containing(df_2015, "GROSS RENT AS A PERCENTAGE OF HOUSEHOLD INCOME")
df_2016 = delete_columns_containing(df_2016, "GROSS RENT AS A PERCENTAGE OF HOUSEHOLD INCOME")
df_2017 = delete_columns_containing(df_2017, "GROSS RENT AS A PERCENTAGE OF HOUSEHOLD INCOME")
df_2018 = delete_columns_containing(df_2018, "GROSS RENT AS A PERCENTAGE OF HOUSEHOLD INCOME")
df_2019 = delete_columns_containing(df_2019, "GROSS RENT AS A PERCENTAGE OF HOUSEHOLD INCOME")
df_2020 = delete_columns_containing(df_2020, "GROSS RENT AS A PERCENTAGE OF HOUSEHOLD INCOME")
df_2021 = delete_columns_containing(df_2021, "GROSS RENT AS A PERCENTAGE OF HOUSEHOLD INCOME")
df_2022 = delete_columns_containing(df_2022, "GROSS RENT AS A PERCENTAGE OF HOUSEHOLD INCOME")
```

### Owners vs. Renters Over Time

```python
dfs = [df_2011, df_2012, df_2013, df_2014, df_2015, df_2016, df_2017, df_2018, df_2019, df_2020, df_2021, df_2022]

# List of years corresponding to the dataframes
years = list(range(2011, 2023))

# Initialize dictionaries to store the data for each region
data_tenure_02119 = {'Year': [], 'Owner Occupied': [], 'Renter Occupied': []}
data_tenure_02121 = {'Year': [], 'Owner Occupied': [], 'Renter Occupied': []}

# Iterate over each dataframe and extract the relevant data
for year, df in zip(years, dfs):
    if 'Geographic Area Name' in df.columns:
        for index, row in df.iterrows():
            if row['Geographic Area Name'] == 'ZCTA5 02119':
                data_tenure_02119['Year'].append(year)
                owner_occupied = pd.to_numeric(row.get('Owner Occupied', 0), errors='coerce')
                renter_occupied = pd.to_numeric(row.get('Renter Occupied', 0), errors='coerce')
                total = owner_occupied + renter_occupied
                data_tenure_02119['Owner Occupied'].append((owner_occupied / total) * 100 if total > 0 else 0)
                data_tenure_02119['Renter Occupied'].append((renter_occupied / total) * 100 if total > 0 else 0)
            elif row['Geographic Area Name'] == 'ZCTA5 02121':
                data_tenure_02121['Year'].append(year)
                owner_occupied = pd.to_numeric(row.get('Owner Occupied', 0), errors='coerce')
                renter_occupied = pd.to_numeric(row.get('Renter Occupied', 0), errors='coerce')
                total = owner_occupied + renter_occupied
                data_tenure_02121['Owner Occupied'].append((owner_occupied / total) * 100 if total > 0 else 0)
                data_tenure_02121['Renter Occupied'].append((renter_occupied / total) * 100 if total > 0 else 0)

# Convert dictionaries to dataframes
df_tenure_02119 = pd.DataFrame(data_tenure_02119)
df_tenure_02121 = pd.DataFrame(data_tenure_02121)

# Sort the dataframes by Year to ensure correct order
df_tenure_02119 = df_tenure_02119.sort_values(by='Year')
df_tenure_02121 = df_tenure_02121.sort_values(by='Year')

# Plot the data
plt.figure(figsize=(14, 8))

# Plot for ZCTA5 02119
plt.subplot(1, 2, 1)
plt.plot(df_tenure_02119['Year'], df_tenure_02119['Owner Occupied'], label='Owner Occupied')
plt.plot(df_tenure_02119['Year'], df_tenure_02119['Renter Occupied'], label='Renter Occupied')
plt.title('Percentage of Home-Owners vs Renters Over Time for ZCTA5 02119')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.legend()

# Plot for ZCTA5 02121
plt.subplot(1, 2, 2)
plt.plot(df_tenure_02121['Year'], df_tenure_02121['Owner Occupied'], label='Owner Occupied')
plt.plot(df_tenure_02121['Year'], df_tenure_02121['Renter Occupied'], label='Renter Occupied')
plt.title('Percentage of Home-Owners vs Renters Over Time for ZCTA5 02121')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.legend()

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```
